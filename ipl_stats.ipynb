```python
import pandas as pd
import numpy as np
```

```python
df1=pd.read_csv('matches.csv')
df1.head()
```

```python
df1.info()
df1.shape
df1.columns
```

```python
df2=pd.read_csv('deliveries.csv')
df2.head()
```

```python
df2.info()
df2.shape
df2.columns
```

```python
df1.isnull().sum()/len(df1)*100
```

```python
df1=df1.drop('umpire3',axis=1)
```

```python
df1.columns
```

```python
df2.isnull().sum()/len(df2)*100
```

```python
df2['player_dismissed'].fillna('Not Out', inplace=True)
```

```python
df1['city'].fillna('Unknown', inplace=True)

# Drop matches where winner is null (no result)
df1 = df1[df1['winner'].notnull()]
```

```python
df1['team1'].replace({'Delhi Daredevils':'Delhi Capitals'},inplace=True)
df1['team2'].replace({'Delhi Daredevils':'Delhi Capitals'},inplace=True)
df1['winner'].replace({'Delhi Daredevils':'Delhi Capitals'},inplace=True)
```

```python
df2['batting_team'].replace({'Delhi Daredevils':'Delhi Capitals'},inplace=True)
df2['bowling_team'].replace({'Delhi Daredevils':'Delhi Capitals'},inplace=True)
```

```python
df1.head(10)
```

```python
team_wins=df1['winner'].value_counts()
team_wins
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
team_wins.plot(kind='bar')
plt.title("Total Wins by Team in IPL")
plt.xlabel("Teams")
plt.ylabel("Total Wins")
plt.xticks(rotation=75)
plt.savefig("output_graphs/team_wins.png") 
plt.show()
```

```python
total_matches = df1['team1'].value_counts() + df1['team2'].value_counts()
win_percentage = (team_wins / total_matches) * 100
win_percentage.sort_values(ascending=False)
```

```python
toss_match=df1[df1['toss_winner']==df1['winner']]
toss_win_count = toss_match.shape[0]
total_matches = df1.shape[0]

print("Percentage of matches won after winning toss:",
      (toss_win_count/total_matches)*100)
```

```python
decision_wins = df1[df1['toss_winner'] == df1['winner']]['toss_decision'].value_counts()
decision_wins.plot(kind='bar')
plt.title("Toss Decision Success")
plt.show()
```

```python
top_batsmen = df2.groupby('batsman')['total_runs'].sum().sort_values(ascending=False).head(10)
top_batsmen
top_batsmen.plot(kind='barh')
plt.title("Top 10 Run Scorers in IPL")
plt.xlabel("Total Runs")
plt.show()

```

```python
balls_faced = df2[df2['wide_runs']==0].groupby('batsman')['ball'].count()
runs_scored = df2.groupby('batsman')['total_runs'].sum()

strike_rate = (runs_scored / balls_faced) * 100
strike_rate.sort_values(ascending=False).head(10)
```

```python
wickets = df2[df2['player_dismissed'].notnull()]
top_bowlers = wickets['bowler'].value_counts().head(10)
top_bowlers
```

```python
runs_conceded = df2.groupby('bowler')['total_runs'].sum()
balls_bowled = df2.groupby('bowler')['ball'].count()

economy = (runs_conceded / (balls_bowled/6))
economy.sort_values().head(10)
```

```python
chasing_wins = df1[df1['result'] == 'wickets']
chasing_venue = chasing_wins['venue'].value_counts()
```
